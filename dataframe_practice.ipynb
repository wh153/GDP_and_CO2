## 7)

```python
# create a new subset dropping missing values
correct_data = narrowed_data[narrowed_data["inctot"] != 9999999]
```

## 8)

```python
# create a new subset of only employed people
employed_data = correct_data[correct_data["empstat"] == "employed"]
employed_data
```

## 9)

```python
employed_data["educ"].value_counts()
```

## 10)

```python
avginc_12=employed_data[employed_data["educ"] == "grade 12"]["inctot"].mean()

print(f"The average value of income for people whose highest grade level is “grade 12” is: {avginc_12:.2f}.")
```

## 11)

```python
avginc_college=employed_data[employed_data["educ"] == "4 years of college"]["inctot"].mean()
print(f"The average value of income for people who graduated college is: {avginc_college:.2f}.")
```

This indicates that people with a college degree can obtain significantly higher income.

## 12)

```python
avginc_below_highschool=employed_data[
    (employed_data["educ"] == "n/a or no schooling")
    | (employed_data["educ"] == "grade 5, 6, 7, or 8")
    | (employed_data["educ"] == "grade 9")
    | (employed_data["educ"] == "grade 10")
    | (employed_data["educ"] == "grade 11")
]["inctot"].mean()

print (f"The average value of income for people who did not finish high school is: {avginc_below_highschool:.2f}.")
```

This shows that a high school diploma can bring higher income, but not as much as a college deploma does.

## 13)

```python
not_finish_highschool = employed_data[
    (employed_data["educ"] == "n/a or no schooling")
    | (employed_data["educ"] == "grade 5, 6, 7, or 8")
    | (employed_data["educ"] == "grade 9")
    | (employed_data["educ"] == "grade 10")
    | (employed_data["educ"] == "grade 11")
]["inctot"].mean()

complete_9th_grade = employed_data[employed_data["educ"] == "grade 9"]["inctot"].mean()
complete_10th_grade = employed_data[employed_data["educ"] == "grade 10"]["inctot"].mean()
complete_11th_grade = employed_data[employed_data["educ"] == "grade 11"]["inctot"].mean()

complete_highschool = employed_data[employed_data["educ"] == "grade 12"]["inctot"].mean()
complete_college = employed_data[employed_data["educ"] == "4 years of college"]["inctot"].mean()

print(f"Average income for someone who has not finished high school: {not_finish_highschool}\nAverage income for someone who has completed 9th grade: {complete_9th_grade}\nAverage income for someone who has completed 10th grade: {complete_10th_grade}\n\
Average income for someone who has completed 11th grade: {complete_11th_grade}\nAverage income for someone who has completed high school: {complete_highschool}\n\
Average income for someone who has completed college: {complete_college}")
```

## 14) Explanation

Since people who has completed 9th,10th and 11th grade are all regarded as 'not finishing high school', there's no big difference in terms of knowledge and academic status between them. People who has completed 9th grade might even go to work ealier, has more experience and working skills, leading to a higher average income compared to those who has completed 10th and 11th grade.
